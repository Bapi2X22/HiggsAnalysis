```python
# Convert to NumPy if starting from pandas DataFrame
reco_pt = df["RecoPhoton_pt_1"].to_numpy()
gen_pt  = df["GenPhoton_pt_1"].to_numpy()

# Filter both arrays for values > 0
mask = (reco_pt > 0) & (gen_pt > 0)
filtered_reco_pt = reco_pt[mask]
filtered_gen_pt  = gen_pt[mask]

# Define 2D histogram: 100 bins in each direction, range 0200
hist2d = ROOT.TH2F("h2d", "Reco vs Gen Photon pT;GenPhoton_pt_1;RecoPhoton_pt_1",
                   100, 0, 100,   # X axis (Gen)
                   100, 0, 100)   # Y axis (Reco)

# Fill the histogram
for gen, reco in zip(filtered_gen_pt, filtered_reco_pt):
    hist2d.Fill(gen, reco)

# Draw the histogram
canvas = ROOT.TCanvas("canvas", "Canvas", 800, 700)
canvas.SetRightMargin(0.15)  # leave space for color palette
hist2d.Draw("COLZ")

canvas.Draw()

canvas.SaveAs("Reco_vs_GenPhoton_pt_1.png")
```
