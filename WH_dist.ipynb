```python
def dR(Pho_gen_eta, Pho_gen_phi, Pho_reco_eta, Pho_reco_phi):
    """
    Calculate the delta R between photons and b quarks.
    """
    d_eta = Pho_gen_eta - Pho_reco_eta
    d_phi = Pho_gen_phi - Pho_reco_phi
    
    return np.sqrt(d_eta**2 + d_phi**2)
```

```python
import numpy as np

# Convert to NumPy arrays if not already
pt1 = ak.to_numpy(pho_from_a_pt_1)
pt2 = ak.to_numpy(pho_from_a_pt_2)
eta1 = ak.to_numpy(pho_from_a_eta_1)
eta2 = ak.to_numpy(pho_from_a_eta_2)
phi1 = ak.to_numpy(pho_from_a_phi_1)
phi2 = ak.to_numpy(pho_from_a_phi_2)

# Compute invariant mass squared
delta_eta = eta1 - eta2
delta_phi = phi1 - phi2

# Ensure delta_phi is within [-, ] to avoid issues
delta_phi = (delta_phi + np.pi) % (2 * np.pi) - np.pi

m2 = 2 * pt1 * pt2 * (np.cosh(delta_eta) - np.cos(delta_phi))

# Set negative values (due to numerical errors) to zero
m2 = np.maximum(m2, 0)

# Compute m_gg
m_gg = np.sqrt(m2)
```

```python
import awkward as ak
import numpy as np

# Step 1: Select valid events where both photon 1 and photon 2 exist
mask_valid = ~ak.is_none(selected_photon_1) & ~ak.is_none(selected_photon_2)

# Step 2: Apply the mask to all components
pt1 = selected_photon_1[mask_valid].pt
pt2 = selected_photon_2[mask_valid].pt
eta1 = selected_photon_1[mask_valid].eta
eta2 = selected_photon_2[mask_valid].eta
phi1 = selected_photon_1[mask_valid].phi
phi2 = selected_photon_2[mask_valid].phi

# Step 3: Compute invariant mass
delta_eta = eta1 - eta2
delta_phi = phi1 - phi2
delta_phi = (delta_phi + np.pi) % (2 * np.pi) - np.pi  # wrap to [-, ]

m2 = 2 * pt1 * pt2 * (np.cosh(delta_eta) - np.cos(delta_phi))
m2 = ak.where(m2 > 0, m2, 0)  # prevent small negative values from sqrt error

m_gg_reco = np.sqrt(ak.to_numpy(m2))
```
